## Arrests 

In this notebook, we will focus on analyzing the total number of arrests by neighborhood in Pittsburgh for the year 2020. By examining and visualizing this data, we aim to identify the neighborhoods with the highest and lowest number of arrests during that year. We will utilize Python programming and data visualization libraries to process the dataset, filter the data to include only arrests from 2020, and create informative visualizations, such as bar graphs and tables, to present our findings effectively.

By analyzing the distribution of arrests by neighborhood, we can identify areas with higher law enforcement activity and potential public safety concerns, as well as areas with lower arrest rates, which may be indicative of safer communities.

```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
```

```python
#Setting up table so that it shows all data instead of getting cut off 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# import dataset
arrests = pd.read_csv("arrests.csv")
arrests_2020 = arrests[arrests['ARRESTTIME'].str.contains('2020', na=False)]
apn = arrests_2020['INCIDENTNEIGHBORHOOD'].value_counts()
#Sort so lowest are at the top
apn_sorted = apn.sort_values(ascending=True)
```

```python
#Setting up bar graph
plt.figure(figsize=(20, 8))
apn_sorted.plot(kind='bar', color='salmon')
plt.xlabel('Neighborhood')
plt.ylabel('Total Arrests')
plt.title('Total Arrests in 2020 by Neighborhood in Pittsburgh')
plt.grid(axis='y')
plt.yticks(rotation=0, ha='right')
plt.show()
```

The bar graph above displays the total number of arrests by neighborhood in Pittsburgh for the year 2020, with neighborhoods sorted in descending order based on arrest counts. In this visualization, the neighborhoods are represented on the x-axis, allowing for easier comparison and identification of areas with higher arrest rates in 2020. The neighborhoods with the tallest bars on the graph indicate higher numbers of arrests during that year, potentially identifying them as areas with increased law enforcement activity or public safety concerns.

The table below presents the total arrests by neighborhood for the year 2020 in a comprehensive format, listing all neighborhoods without truncation. This detailed table provides a complete overview of the arrest counts for each neighborhood in Pittsburgh during 2020, offering valuable insights into the distribution of law enforcement activity across the city for that specific year. 

```python
#Setting up table 
table_data = pd.DataFrame(apn_sorted)
table_data.columns = ['Total Arrests in 2020']
table_data.index.name = 'Neighborhood'
print("\nTable: Total Arrests by Neighborhood in Pittsburgh")
print(table_data)
```
