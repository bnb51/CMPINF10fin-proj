## Firearm Seizures 

In this notebook, we will explore the total number of firearm seizures across various neighborhoods in Pittsburgh. By analyzing and visualizing this data, we aim to identify the neighborhoods with the lowest number of firearm seizures, which may be indicative of safer areas within the city. We will utilize Python programming and data visualization libraries to process the dataset and create informative visualizations, such as bar graphs, to present our findings effectively.



```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
```

```python
#Allowing table at bottom to show all data instead of getting cut off
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# import dataset
seizures = pd.read_csv("firearm_seizures.csv")
# filter to 2020 data
seizures_2020 = seizures[seizures['year'] == 2020]
total_seizures = seizures.groupby("neighborhood").sum()['total_count']
# Sorting so lowest are at the top
total_seizures_sorted = total_seizures.sort_values(ascending=True)
```

```python
#Setting up bar graph
plt.figure(figsize=(20, 8))
total_seizures_sorted.plot(kind='bar', color='skyblue')
plt.xlabel('Neighborhood')
plt.ylabel('Total Seizures')
plt.title('Total Firearm Seizures by Neighborhood in Pittsburgh')
plt.grid(axis='y')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The bar graph above displays the total number of firearm seizures by neighborhood in Pittsburgh, with neighborhoods sorted in descending order based on seizure counts. In this visualization, the neighborhoods are represented on the x-axis, allowing for easier comparison and identification of areas with lower firearm-related incidents. The neighborhoods with the shortest bars on the graph indicate lower numbers of firearm seizures, potentially identifying them as safer areas within the city.

Additionally, the table below presents the same data in a tabular format, providing a detailed view of the total firearm seizures by neighborhood. This table complements the bar graph by offering a clear and concise summary of the seizure counts for each neighborhood in Pittsburgh.

```python
#Setting up table
table_data = pd.DataFrame(total_seizures_sorted)
table_data.columns = ['Total Seizures']
table_data.index.name = 'Neighborhood'
print("\nTable: Total Firearm Seizures by Neighborhood in Pittsburgh")
print(table_data)
```
